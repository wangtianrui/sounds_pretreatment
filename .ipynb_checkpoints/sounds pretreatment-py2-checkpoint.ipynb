```python
def plot_sound(audio):
    x_ = np.arange(0,len(audio))
    plt.plot(x_,audio)
    plt.show()
    
def scat_array(array):
    x_ = np.arange(0,len(array))
    plt.scatter(x_,array)
    plt.show()
plot_sound(audio)
```

```python
import sounddevice as sd
```

```python
def play_sound(sound_array,freq):
    """
    play a sound 
    """
    sd.play(sound_array,sampling_freq)
```

```python
def pre_emphasis(sound_array,u):
    """
    u:预加重系数  coefficient
    """
    temp = np.append(sound_array[0], sound_array[1:] - u * sound_array[:-1])
    print(temp)
    return temp
```

```python
#分帧、加窗
def enframe(signal,nw,inc,window_func=None):
    """
    signal:原始信号
    nw:帧长
    inc:相邻帧间隔,步长
    window_func:窗函数
    """
    signal_length = len(signal)
    if signal_length < nw:
        # nf为帧的数量
        nf = 1
    else:
        # np.ceil 计算大于该值的最小整数
        nf = int(np.ceil((1.0*signal_length-nw+inc)/inc))
    pad_length = int((nf-1)*inc+nw) #所有帧的帧长之和
    zeros = np.zeros((pad_length-signal_length,)) # 补0
    pad_signal = np.concatenate((signal,zeros))
    indices = np.tile(np.arange(0,nw),(nf,1))+np.tile(np.arange(0,nf*inc,inc),(nw,1)).T
    indices = np.array(indices,dtype=np.int32)
    frames = pad_signal[indices]
    if window_func is not None:
        win = np.tile(window_func,(nf,1))
        return frames*win
    return frames
```

```python
frames = enframe(pre_emphasised,512,256)
print(frames)
```

```python
plot_sound(pre_emphasised)
plot_sound(frames.flatten())
play_sound(frames.flatten(),len(frames.flatten()))
```

```python
import scipy.signal as signal
hanning_frames = enframe(pre_emphasised,512,256,signal.windows.hann(512))
plot_sound(pre_emphasised)
plot_sound(hanning_frames.flatten())
play_sound(hanning_frames.flatten(),len(hanning_frames.flatten()))
```

### 倒谱特征——LPCC

```python
import scikits.talkbox as tb
from scikits.talkbox.linpred import levinson_lpc
```

```python
def lpc_to_cc(lpc,n_lpc,n_lpcc): 
    lpcc = np.zeros(n_lpcc)
    lpcc[0] = lpc[0]
    for n in range(1, n_lpc):
        lpcc[n] = lpc[n]
        for l in range(0, n):
            lpcc[n] += lpc[l] * lpcc[n - l - 1] * (n - l) / (n + 1)
    for n in range(n_lpc, n_lpcc):
        lpcc[n] = 0
        for l in range(0, n_lpc):
            lpcc[n] += lpc[l] * lpcc[n - l] * (n - l) / n
    return -lpcc[1:]
```

```python
def lpcc(frames,n_lpc,n_lpcc):
    """
    n_lpc:lpc阶数
    n_lpcc:lpcc阶数
    """
    lpcs=[]
    lpccs = []
    for frame in frames:
        lpc = levinson_lpc.lpc(frame, n_lpc)[0]
        lpcc = lpc_to_cc(lpc,n_lpc,n_lpcc)
        lpcs.append(lpc)
        lpccs.append(lpcc)
    return lpcs , lpccs
```

```python
lpc_ , lpcc_ = lpcc(hanning_frames,3,5)
```

#### 作用
* 首先是LPC，LPC将高维语音变量降维了
* LPCC是在LPC基础上，利用倒谱分析，分析出激励源与声道相应的函数

#### 问题
* 无

#### 发现
* LPCC利用线性拟合的方式，用相应的参数来表示信号，降维过于厉害
* 尝试将维数增大，发现变化不大。
* 可视化意义不大（参数在时域上联系不大）

#### 个人理解
* 维数增大是将阶数提升，但是系数还是会很小的，所以没有意义
* 降维过于厉害，不适合投入强组合分类器进行分类
* 易造成潜在因素的损失
* 阶数不易过高，阶数与参数的值的大小成反比
